```python
#Matrix Multiplication 
tensor1=torch.tensor([[4,3],
                      [2,1],
                      [0,2]])
tensor2=torch.tensor([[3,7],
                      [4,8],
                      [1,2]])
torch.matmul(tensor1.T,tensor2)
```

```python
#Matrix multiplication by hand
tensor=torch.tensor([[45,7],
                     [78,6]])
value=0
for i in range(len(tensor)):
    value += tensor[i] * tensor[i]
print(value)
```

```python
#Two main rule of matrix multiplication

mat=torch.tensor([[5,8,7,5,3],
                 [5,8,7,5,3]])
print(mat.shape)
print(mat.ndim)
```

```python
arr=([4,8,9])
print(arr[:5])
```

```python
%%time
#Multiplying some matrix

mat_1=torch.tensor([[7,8,9],
                    [8,9,7],
                    [8,1,10]])
mat_2=torch.tensor([[2,0,12],
                    [5,1,11],
                    [100,3,2]])

print(torch.matmul(mat_1,mat_2))
print(mat_1@mat_2)


```

```python
#Finding min,max,mean,sum etc

tensor1=torch.tensor([[7,8,9],
                    [8,19,7],
                    [8,1,10]])
rang=torch.arange(10,100,10)
print(torch.min(rang))
print(torch.max(tensor1))
print(torch.mean(rang.type(torch.float64)))
print(torch.sum(tensor1))
print(torch.argmin(rang))
```

```python
#Reshape,squeeze,unsqueeze

reshape=rang.reshape(9)
squeeze=tensor1.squeeze()
print(squeeze.shape)

print(reshape.shape,reshape)
```
