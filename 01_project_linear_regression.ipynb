```python
#Basic things require for the model to be build 

import torch
from torch import nn
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
#The weights and bias for the linear regression formula
weight=0.5
bias=0.3

#Defining the input of the model in the linear regression model
x=torch.arange(0,1,0.02).unsqueeze(dim=1)

y=weight*x+bias
```

```python
#Creating the data set to model get trained on 
train_split=int(0.8*len(x))
x_train=x[:train_split]
x_label=y[:train_split]

y_test=x[train_split:]
y_label=y[train_split:]

print(f"The model training set are {x_train} and the preds are {x_label}")
print(f"The model test sets are {y_test} and the preds are {y_label}")
```

```python
#Plot the prediction 

def plot_predictions(train_data=x_train,
                    train_label=x_label,
                    test_data=y_test,
                    test_label=y_label,
                    prediction=None):
    #Plotting the training graph
    plt.scatter(train_data,train_label,c="b",label="Training data")
    plt.scatter(test_data,test_label,c="r",label="Test data")
    plt.legend(prop={"size":14})
    plt.grid
```

```python
plot_predictions()
```
